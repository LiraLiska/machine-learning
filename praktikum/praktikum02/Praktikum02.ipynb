```python
# buat baca file csv pakai pandass
import pandas as pd

df = pd.read_csv(path + '/data/500_Person_Gender_Height_Weight_Index.csv')
df
```

```python
#ini buat cari info data(ada tipe data, non null count data,dan nama kolom)
df.info()
```

```python
#buat itung mean semua kolom numerik
df['Height'].mean()
```

```python
# buat itung median semua kolom numerik
df['Height'].median()
```

```python
#buat cari modus(hati hati karena bisa lebih dari satu)
df['Height'].mode()
```

```python
# menghitung variasi & standard deviasi
df.var(numeric_only=True)
```

```python
# menghitung standard deviansi
df.std(numeric_only=True)
```

```python
# menghitung quartil pertama (Q1)
q1 = df['Height'].quantile(0.25)
print("Q1 : ", q1)

# menghitung quartil pertama (Q3)
q3 = df['Height'].quantile(0.75)
print("Q3 : ", q3)

# menghitung iqr
iqr = q3 - q1
print("IQR : ", iqr)
```

```python
df.describe()
```

```python
# menghitung matriks korelasi untuk semua kolom numerik
correlation_matrix = df.corr(numeric_only=True)

# menampilkan matriks korelasi
print("Matriks Korelasi : ")
print(correlation_matrix)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# ambil data height
data_height = df["Height"]

#buat histogram
n, bins, patches = plt.hist(data_height, bins=5, color='pink', edgecolor='black')

#tambahkan label
plt.title('Histogram nilai')
plt.xlabel('Height')
plt.ylabel('Frekuensi')

# tampilkan rentang frekuensi di sumbu y
bin_centers = 0.5 * (bins[:-1] + bins[1:])
plt.xticks(bin_centers, [f'{bins[i]:.0f}-{bins[i+1]:.0f}' for i in range(len(bins)-1)])

#tampilkan plot
plt.show()
```
